```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.utils import resample


# Загрузка датасета
file_path = r'C:\p\CSA_monitorv5_processed.csv'
data = pd.read_csv(file_path)

# Балансировка target_value до 50/50
data_0 = data[data['target_value'] == 0]
data_1 = data[data['target_value'] == 1]

# Upsample minority class
data_1_upsampled = resample(data_1, replace=True, n_samples=len(data_0), random_state=123)

# Объединение классов после upsampling
data_balanced = pd.concat([data_0, data_1_upsampled])

# Создание дополнительных признаков Probability_X
for i in range(1, 10):
    data_balanced[f'Probability_{i}'] = (data_balanced['Probability'] >= i).astype(int)

#Разделение данных и обучение модели

# Подготовка признаков и целевой переменной
X = data_balanced.drop(['id', 'target_value', 'Probability'], axis=1)
y = data_balanced['target_value']

# Разделение данных на обучающую и тестовую выборки
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Обучение модели RandomForest
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

```

```python
from joblib import dump

# Задаем путь для сохранения модели на Google Drive
model_path = 'c:/p/model_csa_riskv1.joblib'

# Сохраняем модель
dump(model, model_path)

print(f'Модель сохранена по пути: {model_path}')

```
